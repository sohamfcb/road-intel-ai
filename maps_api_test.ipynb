```python
import googlemaps
from dotenv import load_dotenv
import os

load_dotenv()

maps_api_key=os.getenv("GOOGLE_MAPS_API_KEY")
```

## Python Client for Google Maps Services
Check out the following url to view the documentation of the `googlemaps` python library.

[Click here](https://googlemaps.github.io/google-maps-services-python/docs/index.html)

```python
#geocode api
gmaps=googlemaps.Client(key=maps_api_key)

geocode_result=gmaps.geocode("Kolkata, India")
```

```python
geocode_result[0]['geometry']
```

```python
geocode_result[0]['geometry']['location'],type(geocode_result[0]['geometry']['location'])
```

```python
# directions api
directions=gmaps.directions("Kolkata, India","Delhi, India", mode='driving')
```

```python
directions
```

```python
legs=directions[0]['legs'][0]
```

```python
print(legs['steps'][0]['html_instructions'])
```

```python
for step in legs['steps']:
    print(step['html_instructions'])
    print(step['distance']['text'])
    print(step['duration']['text'])
```

```python
from bs4 import BeautifulSoup

def clean_html(html_text):
    return BeautifulSoup(html_text, "html.parser").get_text()

for step in legs['steps']:
    print(clean_html(step['html_instructions']))
    print()
    print(step['distance']['text'])
    print(step['duration']['text'])
    print()
    print()
```

```python
query = "Taj Mahal"
suggestions = gmaps.places_autocomplete(query)

for s in suggestions:
    print(s["description"])  # Suggested place names
```

```python
search = gmaps.places(query="best cafes in Kolkata")

for place in search["results"]:
    print(place["name"], "-", place["formatted_address"])
    print()
```
